```python
import pandas as pd 
import numpy as np
```

```python
# Load the dataset
df = pd.read_csv("data/WB_WDI_EG_ELC_ACCS_ZS.csv")
```

```python
# View the dataset
df.head(10)
```

```python
df= df[['REF_AREA','REF_AREA_LABEL','TIME_PERIOD','OBS_VALUE']]
df.columns=['countryCode','countryName','year','electricityAccess']
```

```python
df.info()
```

```python
df.head(20)
```

```python
# Ensure 0 ≤ electricityAccess ≤ 100
invalid_low = df[df['electricityAccess'] < 0]
invalid_high = df[df['electricityAccess'] > 100]

print("Below 0:", len(invalid_low))
print("Above 100:", len(invalid_high))
```

```python
df_b42000 =df[df['year'] < 2000]
df_after2000 = df[df['year'] >= 2000]
# count the number of countries before 2000
print("Records :", len(df['countryCode'].unique()))
print("Records before 2000:", len(df_b42000['countryCode'].unique()))
print("Records after 2000:", len(df_after2000['countryCode'].unique()))
```

```python
# Remove records before 2000
df = df[df['year'] >= 2000]
```

```python
aggregate_codes = {
    'AFE','AFW','ARB','CEB','CSS','EAR','EAS','EAP','EMU','ECA','ECS','EUU',
    'FCS','HIC','IBD','HPC','IDA','IBT','IDB','IDX','LAC','LDC','LMY','LMC',
    'LTE','MEA','MNA','MIC','LCN','NAC','LIC','OED','OSS','PRE','PSS','PST',
    'SSA','TEA','TEC','TSA','TMN','TSS','UMC','SAS','WLD','SSF','SST','TLA'
}
```

```python
# Countries
df_countries = df[~df['countryCode'].isin(aggregate_codes)]

# Aggregates
df_groups = df[df['countryCode'].isin(aggregate_codes)]
```

```python
df_countries.head(50)
```

```python
df_groups.head(10)
```

```python
region_df = pd.read_csv("data/country_region.csv")
```

```python
# Merge region information with the countries dataset
df_countries = df_countries.merge(region_df, on='countryCode', how='left')
```

```python
df_countries.info()
```

```python
missing = df_countries[df_countries['region'].isna()]
print(missing['countryCode'].unique())
```

```python
missing.head(20)
```

```python
# Drop records with missing region information
df_countries = df_countries.dropna(subset=['region'])
df_countries.info()
```

```python
# Save the cleaned dataset
df_countries.to_csv("data/cleaned_electricity_access.csv", index=False)
```
